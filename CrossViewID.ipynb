```python
# Cell 1: Imports & Setup
import cv2
import torch
import numpy as np
from pathlib import Path
from utils.detection import run_detection
from utils.tracking import track_players
from utils.matching import match_players_across_views

BROADCAST_PATH = 'broadcast.mp4'
TACTICAM_PATH = 'tacticam.mp4'
MODEL_PATH = 'best.pt'
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
OUTPUT_DIR = Path("output")
OUTPUT_DIR.mkdir(exist_ok=True)
```

```python
broadcast_detections = run_detection(BROADCAST_PATH, MODEL_PATH, device=DEVICE)
tacticam_detections = run_detection(TACTICAM_PATH, MODEL_PATH, device=DEVICE)
```

```python
# Cell 3: Track players using simple IoU + appearance features
broadcast_tracks = track_players(broadcast_detections)
tacticam_tracks = track_players(tacticam_detections)
```

```python
# Cell 4: Match players across camera views
player_mapping = match_players_across_views(broadcast_tracks, tacticam_tracks)

# Print mapping results
print("Cross-camera player ID mapping:")
for tacticam_id, broadcast_id in player_mapping.items():
    print(f"Tacticam ID {tacticam_id} --> Broadcast ID {broadcast_id}")
```
